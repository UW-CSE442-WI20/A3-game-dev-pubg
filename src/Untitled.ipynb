```python
import json

with open('person_fixed.json') as json_file:
    data = json.load(json_file)
```

```python
data[1]
```

```python
for i, year_entry in enumerate(data):
    for j, entity in enumerate(year_entry["entities"]):
        total_user_score = 0
        total_critic_score = 0
        for k, game in enumerate(entity["game"]):
            total_user_score += game["user_score"]
            total_critic_score  += game["critic_score"]
        average_user_score = int(total_user_score / len(entity["game"]) * 10) / 10 if len(entity["game"]) != 0 else 0
        average_critic_score = int(total_critic_score / len(entity["game"]) * 10) / 10 if len(entity["game"]) != 0 else 0
        data[i]["entities"][j]["average_user_score"] = average_user_score
        data[i]["entities"][j]["average_critic_score"] = average_critic_score
```

```python
data[1]
```

```python
with open('person_fixed_with_score.json', 'w') as outfile:
    json.dump(data, outfile)
```
